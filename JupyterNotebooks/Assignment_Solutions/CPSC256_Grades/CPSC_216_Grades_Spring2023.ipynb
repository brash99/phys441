```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dark_mode = True
# plotting choices for dark-mode
if (dark_mode):
    from jupyterthemes import jtplot
    jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)
    linecolor = 'w'
else:
    linecolor = 'k'
    
df = pd.read_csv('CPSC_216_Grades_Spring2023.csv')
df.info()

df_project = pd.read_csv('CPSC_216_Spring_2023_FinalProject.csv')
df_project.info()

df_project
```

```python
df_hw = df.iloc[:, [6]]
df['HW'] = df_hw.sum(axis=1)/1.0

df_part = df.iloc[:, [4]]
df['Participation'] = df_part.sum(axis=1)/1.0

df_challenge = df.iloc[:, [5]]
df['Challenge'] = df_challenge.sum(axis=1)/1.0

df['Midterm'] = df['5.13 - Lab (10)']*0.3333+df['5.14 - Lab (20)']*0.6667

df_proj = df_project.iloc[:,[6]]
df['Project'] = df_proj.sum(axis=1)/1.0
df['ID'] = df_project['ID']

df['Term Numerical Grade'] = 1.33333*(df['Midterm']/100.0*10.0 + df['HW']/100.0*30.0 + df['Challenge']/100.0*10.0 + df['Participation']/100.0*25.0)
df['Total Numerical Grade'] = df['Term Numerical Grade']*0.75+df['Project']*0.25
df['Adjusted Numerical Grade'] = df['Total Numerical Grade']*1.10

df_final = df.iloc[:, [248,0,1,243,244,245,246,247,249,250,251]]

df_final.info()
df_final
```

```python
df_final['Final Letter Grade'] = pd.cut(df_final['Adjusted Numerical Grade'], bins=[0, 49.5, 53.0, 57.0, 59.5, 63.0, 67.0, 69.5, 73.0, 77.0, 79.5, 86.5, 100], labels= ['F', 'D-', 'D', 'D+', 'C-', 'C', 'C+',  'B-', 'B', 'B+', 'A-', 'A'])
df_final
```

```python
df_final2 = df.iloc[:, [248,243,244,245,246,247,249,250,251]]
df_final2['Final Letter Grade'] = pd.cut(df_final2['Adjusted Numerical Grade'], bins=[0, 49.5, 53.0, 57.0, 59.5, 63.0, 67.0, 69.5, 73.0, 77.0, 79.5, 86.5, 100], labels= ['F', 'D-', 'D', 'D+', 'C-', 'C', 'C+',  'B-', 'B', 'B+', 'A-', 'A'])

df_final2.info()
df_final2
```
