```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
    
df = pd.read_csv('CNUCPSC250BrashSummer2023_assignment_report_2023-07-01_0951.csv')
df.info()

df_classlist_250 = pd.read_csv('CPSC250_ClassList.csv')
df_classlist_250L = pd.read_csv('CPSC250L_ClassList.csv')
```

```python
df_final_exam = pd.read_csv('CNUCPSC250BrashSummer2023_CPSC250_-_Final_Exam_report_2023-06-30_1333.csv')
df_final_exam.info()
```

```python
df_hw = df.iloc[:, [3, 4, 6, 7, 9, 11, 12]]
df['HW'] = df_hw.sum(axis=1)/7.0

df_test1 = df.iloc[:, [5]]
df['Test1'] = df_test1.sum(axis=1)/1.0

df_test2 = df.iloc[:, [8]]
df['Test2'] = df_test2.sum(axis=1)/1.0

df_test3 = df.iloc[:, [10]]
df['Test3'] = df_test2.sum(axis=1)/1.0

df_finalexam = df_final_exam.iloc[:, [4]]
df['Exam'] = df_finalexam.sum(axis=1)/1.0

df['Final Numerical Grade'] = (df['HW']/100.0*45.0 + df['Test1']/100.0*10.0 + df['Test2']/100.0*10.0 + df['Test3']/100.0*10.0 + df['Exam']/100.0*25.0)/100.0*100.0
df.info()
```

```python
df_classlist = pd.concat([df_classlist_250,df_classlist_250L]).drop_duplicates().reset_index(drop=True)
df_classlist = df_classlist.sort_values(by=['Name'])
df_classlist = df_classlist.reset_index()
df_classlist
```

```python
df_final = df.iloc[:, [0,1,21,22,23,24,25,26]]
df_final['Final Letter Grade'] = pd.cut(df_final['Final Numerical Grade'], bins=[0, 49.5, 53.0, 57.0, 59.5, 63.0, 67.0, 69.5, 73.0, 77.0, 79.5, 86.5, 100], labels= ['F', 'D-', 'D', 'D+', 'C-', 'C', 'C+',  'B-', 'B', 'B+', 'A-', 'A'])
df_final = df_final.join(df_classlist['ID'])
df_final = df_final.join(df_classlist['Name'])
df_final = df_final.drop([1,6,8])
df_final
```

```python
df_report = df_final.iloc[:, [9,2,3,4,5,6,7,8]]
df_report
```
